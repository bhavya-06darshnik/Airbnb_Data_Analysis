LOADING DATA

```python
listings=pd.read_csv('data/listings.csv')
listings.head()
```

# DATA CLEANING

Dropping unnecsseary features, i.e. not important for us.

```python
misc_features=[
    'listing_url',
    'scrape_id',
    'name',
    'description',
    'neighborhood_overview',
    'picture_url',
    'host_url',
    'host_location',
    'host_about',
    'host_thumbnail_url',
    'host_picture_url',
    'neighbourhood',
    'neighbourhood_group_cleansed',
    'calendar_updated',
    'calendar_last_scraped',
    'license'
]

listings=listings.drop(misc_features,axis=1)
```

```python
listings.head()
```

# DATA ADJUSTMENTS

Data adjusments is meant to make data less complex and messy, by not removing but only adjusting as per requirements.



```python
import re
regex=re.compile('[a-z A-Z]')


listings['bathrooms']=listings['bathroooms_text'].fillna('NaN').str.replace(regex, r"\0", regex=True).str.strip('\x00 - ').astype('string').replace('','NaN').astype('float64')
listings['host-verifications']=listings['host-verifications'].map(lambda x:len([_x.strip() for _x in eval(x)]))
listings['amenities']=listings['amenities'].map(lambda x:len([x for x in eval(x)]))
listings['price']=listings['price'].astype('string').str.strip('$').str.replace('%','').astype('float64')
```
